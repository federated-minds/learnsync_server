```python
import pandas as pd
df=pd.read_csv("mock.csv")
```

```python
df
```

```python
# Define weights for each parameter
weight_time_spent = 0.4
weight_quiz_scores = 0.4
weight_quiz_attempts = -0.2  # Negative weight for quiz attempts since low attempts should give a higher score

# Function to calculate the composite score for each row
def calculate_composite_score(row):
    return (
        row['time_spent'] * weight_time_spent +
        row['quiz_scores'] * weight_quiz_scores +
        row['quiz_attempts'] * weight_quiz_attempts
    )

# Apply the function to create a new column 'composite_score'
df['composite_score'] = df.apply(calculate_composite_score, axis=1)

# Normalize the composite scores to be in the range [0, 100]
min_score = df['composite_score'].min()
max_score = df['composite_score'].max()

df['composite_score_normalized'] = (100 * (df['composite_score'] - min_score) / (max_score - min_score)).round().astype(int)

# Print or save the result
final_df=pd.DataFrame(df[['time_spent', 'quiz_scores', 'quiz_attempts', 'composite_score_normalized']])
final_df
```

```python
categories = ['Poor', 'Fair', 'Good']
score_ranges = [0, 33, 66, 100]

# Categorize individuals based on their composite scores
df['performance_category'] = pd.cut(df['composite_score_normalized'], bins=score_ranges, labels=categories, include_lowest=True)

# Print or save the result
df.head(20)
final_df1=pd.DataFrame(df)
final_df1
```

```python
performance_mapping = {'Poor': 0, 'Fair': 1, 'Good': 2}

df['performance_numeric'] = df['performance_category'].map(performance_mapping)
final_df2=pd.DataFrame(df)
final_df2
```

```python
df=final_df2.drop("composite_score",axis=1)
df
```

```python
df1=final_df2.drop("performance_category",axis=1)
df1
```

```python
df2=df1.drop("composite_score",axis=1)
df2
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


# Define features (X) and target variable (y)
X = df2[['course_id', 'time_spent', 'quiz_scores', 'quiz_attempts']]
y = df2['performance_numeric']

# Assign minimal weight to course_id
X['course_id'] *= 0.001

# Normalize the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Initialize classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC()
}

# Train and evaluate each classifier
for clf_name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{clf_name} Accuracy: {accuracy:.2f}')
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

X = df2[['course_id', 'time_spent', 'quiz_scores', 'quiz_attempts']]
y = df2['performance_numeric']

# Assign minimal weight to course_id
X['course_id'] *= 0.001

# Normalize the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Train a logistic regression model
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(X_normalized, y)

# Make predictions for the provided input
input_features = pd.DataFrame({'course_id': [1], 'time_spent': [40], 'quiz_scores': [90], 'quiz_attempts': [2]})
input_features['course_id'] *= 0.001  # Assign minimal weight to course_id
input_features_normalized = scaler.transform(input_features)

output_prediction = logistic_regression_model.predict(input_features_normalized)

print(f'Predicted Output: {output_prediction[0]}')
```

```python
import pickle
import pandas as pd

# Load the pickled model file
with open('random_forest_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

df = pd.read_csv("mock2.csv")
result = {}

for index, course in df.iterrows():
    # Create a DataFrame from the row
    input_data = pd.DataFrame([{
        "course_id": course["course_id"],
        "time_spent": course["time_spent"],
        "quiz_scores": course["quiz_scores"],
        "quiz_attempts": course["quiz_attempts"]
    }])
    
    # Ensure the input data has the correct data types
    input_data = input_data.astype({"course_id": int, "time_spent": float, "quiz_scores": float, "quiz_attempts": int})
    
    # Predict using the loaded model
    res = loaded_model.predict(input_data)
    
    result[course["course_id"]] = res[0]  # Assuming res is an array with a single prediction value

print(result)
```
